```python
med_corpus = load_dataset("Dqdung205/medical_vietnamese_datasets")
```

```python
med_corpus['train'][:5]
```

```python
import pandas as pd
from collections import defaultdict

# Convert dataset to pandas for easier manipulation
df = pd.DataFrame(med_corpus['train'])

# Group by question and concatenate answers
grouped_data = defaultdict(list)
for _, row in df.iterrows():
    grouped_data[row['question']].append(row['answer'])

# Create new dataset with concatenated answers
merged_data = []
for question, answers in grouped_data.items():
    merged_answer = ''.join(answers)
    merged_data.append({
        'question': question,
        'answer': merged_answer
    })

```

```python
# Convert back to DataFrame to see results
merged_df = pd.DataFrame(merged_data)
print(f"Original dataset size: {len(df)}")
print(f"Merged dataset size: {len(merged_df)}")
print("\nSample merged data:")
merged_df.head()
```

```python
print(merged_df.answer[1])
```

```python
from datasets import Dataset
from datasets import DatasetDict

ds = Dataset.from_pandas(merged_df)

dataset_dict = DatasetDict({
    'train': ds
})

dataset_dict.push_to_hub("quannguyen204/vietnamese-medical-article-corpus")
```
